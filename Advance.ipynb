# 1. Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

x_1 = pd.DataFrame({"var": var, "var_1": var_1})
print(x_1)
sns.lineplot(y = 'var_1', x="var", data=x_1)
plt.show()
```

```python
sns.lineplot(x=var, y=var_1)
plt.show()
```

# 2. Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

vars = sns.load_dataset('penguins')
vars
```

```python
import seaborn as sns
order_1 = ['Adelie', 'Gentoo', 'Chinstrap']
sns.barplot(x=vars.species, y=vars.bill_depth_mm, hue=vars.sex, hue_order=['Female', 'Male'], order=order_1, palette='dark:r', saturation=0.5, err_kws={'color': 'y', 'linewidth': 3}, capsize=0.3, dodge=True, alpha=0.9)
# 'ci' is blacks line on every top of bar value must be in [0 to 100]# 'n_boot' is do something up or down as theird value, # palette='dark:r' ==== color='r' <-- both are same, use any. and here we can use color but use "palette" # "saturation" give it change little bit color # "dodge" method override each other one seame bar upper side another color lower side another color # "alpha" use for light our chart color

plt.show()
```

### horizontal

```python
# when we are create bar plot horizontaly then both of them value must be numeric(number)
sns.barplot(x=vars.bill_length_mm, y=vars.bill_depth_mm, orient='h') # ci is black line on every bar # n boot is do something up or down as theird value
plt.show()
```

```python
y = {'fontsize': 15, 'color': 'b'}
v = sns.heatmap(x, vmin=0, vmax=12, cmap='flare', annot=True, annot_kws=y, linewidths=1, linecolor='y', cbar=True, xticklabels=False, yticklabels=True) 
# vmin & vmax set color size # cmap for give custom color, # annot give value in square # cbar false remove color bar
v.set(xlabel='python', ylabel='laksh')
plt.show()
```

# 8. Strip Plot

# 9. Box Plot
